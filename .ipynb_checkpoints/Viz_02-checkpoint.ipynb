### pandas Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd

temperature = [25.2, 27.4, 22.9, 26.2, 29.5, 33.1, 30.4, 36.1, 34.4, 29.1]
Ice_cream_sales = [236500,357500, 203500, 365200, 446600, 574200, 453200, 675400,598400, 463100]

dict_data = {'t':temperature, 'sales_amt':Ice_cream_sales}
df_ice_cream = pd.DataFrame(dict_data, columns=['t','sales_amt'])

print(df_ice_cream)

df_ice_cream.plot.scatter(x='t', y='sales_amt', grid=True, title='temp & sales_amt')
plt.show()
```

### pandas Bar Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

grade_num = [5, 14, 12, 3]
students = ['A', 'B', 'C', 'D']

df_grade = pd.DataFrame(grade_num, index=students, columns=['Student'])
print(df_grade)

grade_bar = df_grade.plot.bar(grid=True)
grade_bar.set_xlabel("grade")
grade_bar.set_ylabel("number of students")
grade_bar.set_title("bar plot of number of students per grades")         
```

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

math = [76,82,84,83,90,86,85,92,72,71,100,87,81,76,94,78,81,60,79,69,74,87,82,68,79]

df_math = pd.DataFrame(math, columns=['Student'])

math_hist = df_math.plot.hist(bins=8, grid=True)
math_hist.set_xlabel("test score")
math_hist.set_ylabel("frequency")
math_hist.set_title("histogram of math exam")
```

### pandas Pie Chart

```python
import matplotlib.pyplot as plt
import pandas as pd

fruit=['apple','banana','strawberry','orange','grape']
result=[7,6,3,2,2]

df_fruit = pd.Series(result, index=fruit, name='selected number of students')
print(df_fruit)

df_fruit.plot.pie()
```

```python
explode_value = (0.1, 0, 0, 0, 0)
fruit_pie = df_fruit.plot.pie(figsize=(5,5), autopct='%.1f%%', startangle=90,
                             counterclock=False, explode=explode_value, shadow=True, table=True)
fruit_pie.set_ylabel("")  # remove y-axis
fruit_pie.set_title("result")

plt.savefig('C:/test.png', dpi=200)
plt.show()
```
